## One-hot encoding

머신러닝이나 딥러닝 알고리즘은 수치로 된 데이터만 이해할 수 있다.
그래서 기계가 이해할 수 있는 형태로 데이터를 변환해 주어야 하는데 범주형 데이터는 원핫인코딩 형태로 변환해 준다.

```python
#loc gets rows (or columns) with particular labels from the index.
#iloc gets rows (or columns) at particular positions in the index (so it only takes integers).

for_onehot = data.iloc[:,2:].loc[:, data.nunique() >= 3].columns
for_onehot
```

```python
end_onehot = pd.get_dummies(data[for_onehot])
end_onehot
```

```python
# First, drop 'before one hot variables' , and add 'after one hot variables'
data_without_onehot = data.drop(for_onehot, inplace=False, axis=1)

data_final = pd.concat([data_without_onehot, end_onehot], axis=1)
data_final
```
